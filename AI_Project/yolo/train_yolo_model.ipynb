# 학습 이미지 크기가 너무 크기 때문에 사이즈를 줄입니다

```python
image_path = './coco128/images/train/'
label_path = './coco128/labels/train/'

import os
filenames = os.listdir(image_path)

list_files = []
for file in os.listdir(image_path):
    list_files.append(image_path + file)
```

```python
from PIL import Image

for file in list_files:
    img = Image.open(file)
    print(img.size)
    im = img.resize((img.size[0]//4, img.size[1]//4))
    print(im.size)
    im = im.convert('RGB')
    im.save(file)
```
